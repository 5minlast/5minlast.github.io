## Section 1: NumPy 학습

NumPy의 기본적인 배열 생성, 난수 발생, 그리고 효율적인 데이터 처리 방법을 다룹니다.

### 1.1 NumPy 기초 (`20260304-numpy.ipynb` 내용)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# NumPy 배열 생성 예제
a = np.zeros(10)
b = np.ones(10)
c = np.full(10, 5)
d = np.eye(3)
e = np.arange(10)
f = np.linspace(0, 1, 5)

print("zeros:", a)
print("ones:", b)
print("full:", c)
print("eye:\n", d)
print("arange:", e)
print("linspace:", f)
```

### 1.2 난수 생성과 시각화 (`20260304-numpy-random.py` 및 `20260304-numpy-test.py` 내용)

```python
np.random.seed(42)
random_ints = np.random.randint(1, 11, 1000)
print("First 5 random ints:", random_ints[:5])
print("Shape:", random_ints.shape)

# 데이터 수정 및 통계
random_ints[random_ints == 5] = 100
print("Mean after update:", random_ints.mean())
print("Std after update:", random_ints.std())

# 시각화
plt.hist(random_ints, bins=20)
plt.title("Random Integers Distribution")
plt.show()
```

### 1.3 효율적인 NumPy 작업 (`20260304-numpy-time.py` 내용)

```python
# 비효율적인 방법 (append 사용)
def inefficient_way():
    arr = np.array([])
    for i in range(1000):
        arr = np.append(arr, i)
    return arr

# 효율적인 방법 (리스트 사용 후 변환)
def efficient_way():
    lst = []
    for i in range(1000):
        lst.append(i)
    return np.array(lst)

print("Efficient conversion from list is generally faster than np.append in a loop.")
```

## Section 3: 탐색적 자료분석 (EDA)

실제 데이터셋을 활용한 분석 과정을 담고 있습니다.

### 3.1 Tips 데이터셋 분석 (`20260304-project.ipynb` / `.py` 내용)

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips['price_per_person'] = tips['total_bill'] / tips['size']

print("Tips Data Head:\n", tips.head())

# 성별 및 흡연 여부별 팁 비율 요약
stats = tips.groupby(['sex', 'smoker'])[['tip_pct']].describe()
print("\nGrouping Statistics:\n", stats)

# 시각화: 요일별 팁 비율
sns.boxplot(x='day', y='tip_pct', data=tips)
plt.title("Tip Percentage by Day")
plt.show()
```
